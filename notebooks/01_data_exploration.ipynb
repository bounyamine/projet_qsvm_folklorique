# 01 - Exploration des données audio

Ce notebook permet d'explorer les fichiers audio bruts (Gurna vs non-Gurna),
visualiser les formes d'onde et les spectrogrammes, et vérifier la structure
des dossiers dans `data/raw_audio`.


```python
import numpy as np
from pathlib import Path
import librosa
import librosa.display
import matplotlib.pyplot as plt

DATA_DIR = Path('../data/raw_audio/')
list(DATA_DIR.rglob('*.wav'))[:20] + list(DATA_DIR.rglob('*.mp3'))[:20]
```

```python
# Exemple : afficher la forme d'onde et le spectrogramme d'un fichier
example_path = next(iter(DATA_DIR.rglob('*.wav')), None)
example_path = DATA_DIR / 'gurna' / 'gurna_002.wav'
if example_path is not None:
    y, sr = librosa.load(example_path, sr=22050, mono=True)
    plt.figure(figsize=(10, 3))
    librosa.display.waveshow(y, sr=sr)
    plt.title(f'Forme d\'onde - {example_path.name}')
    plt.show()

    D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogramme (dB)')
    plt.show()
```
