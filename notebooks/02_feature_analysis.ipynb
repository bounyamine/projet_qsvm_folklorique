# 02 - Analyse des features audio

Ce notebook charge le fichier HDF5 de features, explore la distribution
des composantes PCA, et visualise quelques projections 2D/3D.


```python
from pathlib import Path
import h5py
import numpy as np
import matplotlib.pyplot as plt

features_path = Path('results/features/extracted_features.h5')

if features_path.exists():
    with h5py.File(features_path, 'r') as f:
        X = np.array(f['X'])
        y = np.array(f['y'])
    X.shape, np.bincount(y)

# Visualisation simple en 2D des deux premi√®res composantes PCA
if features_path.exists():
    plt.figure(figsize=(6, 5))
    for label in np.unique(y):
        idx = y == label
        plt.scatter(X[idx, 0], X[idx, 1], label=f'classe {label}', alpha=0.7)
    plt.xlabel('PC1')
    plt.ylabel('PC2')
    plt.legend()
    plt.title('Projection PCA des segments audio')
    plt.show()

```
